```python
n = int(input("Enter a odd number"))
m = n*3
p= n//2
r= m-7
z= r//2
a=m-3
t = a//2
l = 3
s=1
for i in range(n):
    if i == p:
        for j in range(3):
            for k in range(z):
                if j == 1:
                    print("WELCOME",end='')
                    break
                else:
                    print('-',end='')
        print()
    elif i > p:
        s= s-2
        for q in range(3):
                if q != 1:
                      for i in range(l):
                           print('-',end='')
                else:
                    for d in range(s):
                      print('.|.',end='')
                    
        print()
        l = l+3

    
    else:
        for q in range(3):
                if q != 1:
                      for i in range(t):
                           print('-',end='')
                else:
                    for i in range(s):
                      print('.|.',end='')
                    s=s+2
        print()
        t = t-3
 
    # print(s)


        
```
