# Statistics Practical Implementation
## Measure Of Dispersion
* Variance
* Standard Deviation

```python
ages=[23,24,32,45,12,43,67,45,32,56,32,120]
```

```python
import numpy as np
import statistics as s
```

```python
s.variance(ages)
```

```python
np.var(ages)
```

```python
s.pvariance(ages)
```

```python
import math
math.sqrt(s.pvariance(ages))
```
